### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
import sympy as sym
from sympy import *

x = sym.Symbol('x')

d1 = diff(x**5 + 4*sin(2*x) + cos(3*x+3), x=1) # первая производная
d2 = diff(d1)                                  # вторая производная

print(f"{d1.evalf(subs={x:1})}\n{d2.evalf(subs={x:1})}") # значения первой и второй производных в точке x = 1
```

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
import sympy as sym

x = sym.Symbol('x')

d1 = diff(sin(2*x+1)**5) # производная 1
d2 = diff(d1)            # производная 2

p1 = plot(d1, xlim=[-5,5], show=False, legend=True) # график производной 1
p2 = plot(d2, show=False)              # график производной 2
p1.append(p2[0])                                    # объединяю графики
p1.show()
```
