### Задание 1

Посчитайте частные производные функций:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
### YOUR CODE HERE
import sympy as sym
from sympy import *

x, y = sym.symbols('x, y') # задаю x и y

f1 = 2*x**2*y**3+1/x+y**2*x+7 # функция 1
f2 = x**2*y-sin(x*y)+cos(x**2)+6*y # функция 2

diff(f1, x, y) # дифференцирую ф.1 по х и по у
```

```python
diff(f2, x, y) # дифференцирую ф.2 по х и по у
```

### Задание 2

Вычислите градиент функции:

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$
