### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
# уменьшение в 2 раза
shrink = np.array([ [0.5,0.0,0.0],
                    [0.0,0.5,0.0],
                    [0.0,0.0,0.0]
                ])
B = A@shrink
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
# поворот на 130 градусов
# комментарии ниже пишу для себя
np.pi # число пи
np.radians(1) # 1 радиан = 1/180 пи

# решение
rot = np.deg2rad(130) # задаю угол повопрота в радианах
c, s = np.cos(rot), np.sin(rot) # задаю косинус и синус угла поворота
rotation = np.array([[c, -s],
                     [s, c],
                     [0, 0]
                  ]) # матрица поворота
R = B@rotation
x = R[:,0]
y = R[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

````python
# отразить объект относительно прямой y=x
# несколько понял, задача поменять оси местами
flat = np.array([
                [0, 1, 0],
                [1, 0, 0], 
                [0, 0, 1]
                ])
F = A@flat
x = F[:,0]
y = F[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
````

````python
# т.к. не был уверен в предыдущем решении, смоделировал 2-х мерный объект, убрав ось Z и применил ф-цию flip
C = [
[-100, -100],
[-100, 100],
[100, 100],
[0, 0],
[100, -100],
[-100, -100]
]
C = np.array(C)

C = np.flip(C, axis=1)
x = C[:,0]
y = C[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
````

### Задание 2

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````python
m = np.array([[1, 2],
              [2, 3]
              ])
mval, mvec = np.linalg.eig(m) # получаю собственные вектора и собственные значения
D = np.diag(mval) # создаю диагональную матрицу с собственными значениями по главной диагонали
S = mvec # матрица, столбцы которой являются собственными векторами матрицы m
ST = S.T # её транспонированный вариант
A = np.dot(S, np.dot(D, ST)) # проверка, матрицы сошлись
print(f"{m}\n{A}")
````
